```python
data = pd.read_csv(r"E:\Internship tasks\student\student-mat.csv", encoding='utf-8')
```

```python
# Ensure no extra spaces in column names
data.columns = data.columns.str.strip()
```

```python
# 1. Average score in math (G3)
# Check if 'G3' exists and proceed
if 'G3' in data.columns:
    average_g3 = data['G3'].mean()
    print("\nAverage score in final grade (G3):", average_g3)
else:
    print("'G3' column not found. Available columns are:", data.columns)
```

```python
# 2. Number of students scoring above 15 in final grade (G3)
# Check if 'G3' exists and calculate the number of students scoring above 15
if 'G3' in data.columns:
    students_above_15 = (data['G3'] > 15).sum()
    print("\nNumber of students scoring above 15 in final grade (G3):", students_above_15)
else:
    print("'G3' column not found. Available columns are:", data.columns)
```

```python
# 3. Correlation between study time and final grade
correlation = data['studytime'].corr(data['G3'])
print("\nCorrelation between study time and final grade (G3):", correlation)
```

```python
# 4. Gender with higher average final grade
average_g3_by_gender = data.groupby('sex')['G3'].mean()
print("\nAverage final grade (G3) by gender:")
print(average_g3_by_gender)
```

## Data Visualization

```python
# 1. Histogram of final grades (G3)
plt.figure(figsize=(8, 6))
plt.hist(data['G3'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of Final Grades (G3)')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# 2. Scatter plot between study time and final grade
plt.figure(figsize=(8, 6))
sns.scatterplot(x='studytime', y='G3', data=data, color='red')
plt.title('Scatter Plot: Study Time vs Final Grade')
plt.xlabel('Study Time (hours/week)')
plt.ylabel('Final Grade (G3)')
plt.grid(True)
plt.show()
```

```python
# 3. Bar chart comparing average scores of male and female students
plt.figure(figsize=(8, 6))
average_g3_by_gender.plot(kind='bar', color=['blue', 'pink'], edgecolor='black')
plt.title('Average Final Grade (G3) by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Final Grade (G3)')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.75)
plt.show()
```
